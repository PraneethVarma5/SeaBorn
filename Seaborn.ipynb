```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# '%matplotlib inline' this allows to display the plots directly in notebook instead of opening in new window
```

```python
print(sns.get_dataset_names()) #gives built in datasets in seaborn
```

```python
#we are using car_crashes dataset
df=sns.load_dataset('car_crashes')
print(df)
```

## Distribution plots

### Distrbution plot

```python
#this provides a way to look at univariate distributions(one variable) 
sns.distplot(df['not_distracted'])
# this is histogram plot and the line is called kernal density estimation
```

### Joint plot

```python
sns.jointplot(x='speeding',y='alcohol',data=df)
```

```python
sns.jointplot(x='speeding',y='alcohol',data=df,kind='reg')
#kind = 'reg' gives a regression line across the graph
```

```python
sns.jointplot(x='speeding',y='alcohol',data=df,kind='kde')
sns.jointplot(x='speeding',y='alcohol',data=df,kind='hex')
```

### KDE plot

```python
sns.kdeplot(df['alcohol'])
```

### Pair Plots

```python
#plots relationships across enitre dataframe numerical values
sns.pairplot(df)
```

```python
#another dataset
df1=sns.load_dataset('tips')
df1
sns.pairplot(df1,hue='sex',palette='Blues')
#this gives a pairplot of the entire dataset based on hue sex and gives the graphes of color blue
```

### Rug plots

```python
sns.rugplot(df1['tip'])
#plots a single col of datapoints in a dataframe as sticks
#more dense no of lines means most common and less common is not dense
```

### Styling


```python
sns.set_style('white')#dark,darkgrid,whitegrid,ticks also
plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale=1.5)#paper can replace with talk and poster 
sns.jointplot(x='speeding',y='alcohol',data=df,kind='reg')
sns.despine(left=True,bottom=True)#used to remove the line of the side mentioned
```

## Categorical Plotting

### Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=df1,estimator=np.mean)
#basically this gives the plot based on the mean and this can be changed using estimator=np.median
```

### Count Plot

```python
sns.countplot(x='sex',data=df1)
```

### Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=df1,hue='sex')
```

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=df1,hue='sex') #it is a mix of box and kde plot
```

### Strip plots

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day',y='total_bill',data=df1,jitter=True,hue='sex',dodge=True)#jitter is used to spread the points a bit
#dodge seperates men and women data
```

### Swarm Plot

```python
#sns.violinplot(x='day',y='total_bill',data=df1) 
sns.swarmplot(x='day',y='total_bill',data=df1)#It is a combination of violin plot
```

### Palettes

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=df1,hue='sex',palette='magma')#these palettes changes the color of visual points.
#For more palettes you can visit matplotlib colormaps
plt.legend(loc=0)
```

## Matrix Plot

### HeatMaps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
df_numeric = df.select_dtypes(include=[np.number])#Filter the DataFrame to include only numeric columns for correlation calculation.
mx = df_numeric.corr() #corr means corelation tells u how a variable is influencial to the result
sns.heatmap(mx,annot=True,cmap='Blues')#cmap means color map
```

### Cluster Maps

```python
df3=sns.load_dataset('iris')
df3
```

```python
sns.clustermap(df3)
```

### Pair Grids

```python
iris=sns.PairGrid(df3,hue='species')
iris.map_diag(plt.hist) 
#iris.map_offdiag(plt.scatter) 
iris.map_upper(plt.scatter) 
iris.map_lower(sns.kdeplot) 
```

```python
i=sns.PairGrid(df3,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
i.map(plt.scatter)
i.add_legend()
```

### Facet Grid

```python
fg=sns.FacetGrid(df1,col='time',row='smoker')
fg.map(plt.hist,'total_bill',bins=8)
```

```python
fg=sns.FacetGrid(df1,col='time',row='smoker')
fg.map(plt.scatter,'total_bill','tip')
```

```python
fg=sns.FacetGrid(df1,col='time',hue='smoker',height=4,aspect=1.3)
fg.map(plt.scatter,'total_bill','tip')
```

```python
att_df=sns.load_dataset('attention')
att_df
```

```python
att_fg=sns.FacetGrid(att_df,col='subject',col_wrap=5,height=1.5)
att_fg.map(plt.plot,'solutions','score',marker='.')
```
